```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("greendestination (1) (1).csv")
df.head()
```

```python
print("Dataset Info:")
df.info()
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nAttrition Counts:")
print(df['Attrition'].value_counts())
```

```python
total_employees = len(df)
left_employees = df[df['Attrition'] == 'Yes'].shape[0]
attrition_rate = (left_employees / total_employees) * 100
print(f"\nAttrition Rate: {attrition_rate:.2f}%")
```

```python
df['Attrition'].value_counts().plot.pie(
    autopct='%1.1f%%', 
    startangle=90, 
    colors=["lightgreen", "salmon"]
)
plt.title("Attrition Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title("Age vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title("Years at Company vs Attrition")
plt.show()
```

```python
df_encoded = df.copy()
df_encoded['Attrition'] = df_encoded['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
plt.figure(figsize=(6, 4))
corr = df_encoded[['Attrition', 'Age', 'MonthlyIncome', 'YearsAtCompany']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## üìå Conclusion: Employee Attrition Analysis

- ‚úÖ **Overall Attrition Rate:** 16.12%
- üîç Employees with **lower monthly income** tend to leave more frequently.
- üîç **Newer employees** (with fewer years at the company) are more likely to quit.
- üîç **Younger age groups** show a slightly higher attrition rate compared to older employees.
- üìä Visuals indicate strong patterns between attrition and factors like income, tenure, and age.
- üì¢ **Recommendation for HR:**
  - Focus on increasing employee engagement for those in early years of employment.
  - Consider strategies like better onboarding, mentorship programs, or early performance rewards.
  - Revisit salary structures for junior roles to improve satisfaction and retention.
